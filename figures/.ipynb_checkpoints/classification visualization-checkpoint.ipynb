```python
import numpy as np
import matplotlib.pyplot as plt
import os
import sys

os.environ['KMP_DUPLICATE_LIB_OK'] = 'True' # dead kernel for matplotlib
```

```python
blend = np.load('./parameters/tune_mix_acc.npy')
merge = np.load('./parameters/merge_mix_acc.npy')
orig = np.load('./parameters/orig_mix_acc.npy')
seg = np.load('./parameters/seg_mix_acc.npy')
orig_blur = np.load('./parameters/orig_t_mix_acc.npy')
```

```python
x = np.arange(1, 100+1)
```

```python
plt.plot(x, blend, label='Experiment 1 (Blending)')
plt.plot(x, merge, label='Experiment 2 (Merge)')
plt.plot(x, orig, label='Experiment 3 (Original)')
plt.plot(x, seg, label='Experiment 4 (Segmentation Only)')
plt.plot(x, orig_blur, label='Experiment 5 (Original + Blur)')

plt.legend()
plt.grid(True)

plt.title('Accuracy Analysis')

plt.xlim(0, 100)
plt.ylim(55, 80)
plt.show()
```

```python
blend_loss = np.load('./parameters/tune_mix_loss.npy')
merge_loss = np.load('./parameters/merge_mix_loss.npy')
orig_loss = np.load('./parameters/orig_mix_loss.npy')
seg_loss = np.load('./parameters/seg_mix_loss.npy')
orig_blur_loss = np.load('./parameters/orig_t_mix_loss.npy')
```

```python
plt.figure(figsize=(15,8))

plt.subplot(1, 2, 1)

plt.plot(x, blend, label='Experiment 1 (Blending)')
plt.plot(x, merge, label='Experiment 2 (Merge)')
plt.plot(x, orig, label='Experiment 3 (Original)')
plt.plot(x, seg, label='Experiment 4 (Segmentation Only)')
plt.plot(x, orig_blur, label='Experiment 5 (Original + Blur)')

plt.legend(loc='lower right')
plt.grid(True)

plt.title('Accuracy Analysis')

plt.xlim(0, 100)
plt.ylim(50, 80)
# plt.show()

plt.subplot(1, 2, 2)

plt.plot(x, blend_loss, label='Experiment 1 (Blending)')
plt.plot(x, merge_loss, label='Experiment 2 (Merge)')
plt.plot(x, orig_loss, label='Experiment 3 (Original)')
plt.plot(x, seg_loss, label='Experiment 4 (Segmentation Only)')
plt.plot(x, orig_blur_loss, label='Experiment 5 (Original + Blur)')

plt.legend()
plt.grid(True)

plt.title('Loss Analysis')

plt.xlim(0, 100)
# plt.show()
plt.savefig('./classification visual.png')
```

```python
print(blend.max(), orig_blur.max(), merge.max(), orig.max(), seg.max(), sep='\n')
```
